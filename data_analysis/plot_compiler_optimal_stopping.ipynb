```python
import pandas as pd
import matplotlib.pyplot as plt
from os.path import exists

all_searched_areas = []
complexities = []
for k in [3,6,9,12,15,18,30]:
    searched_areas = []
    for i in range(30,100,1):
        files = [f"search_data_update/bots-2_min-room-size{k}.0/seed-{j}_{i}explored_Exploration.csv" for j in range(50, 60, 1)]
        result = []
        for file in files:
            if exists("../" + file):
                result.append(pd.read_csv("../" + file))
            else:
                print("file doesn't exist:" + file)
        all_files = pd.concat(result)
        # all_files
        map_complexity = all_files.loc[:, 'map_complexity'].mean()
        total_searched_area = all_files.loc[:, 'total_searched_area'].mean()
        searched_areas.append(total_searched_area)
    complexities.append(map_complexity)
    print("completed k=" + str(k))
    all_searched_areas.append(searched_areas)

# searched_areas
# map_complexity
for searched_area in all_searched_areas:
    plt.plot([x for x in range(30,100,1)], searched_area, label=f"Complexity {complexities.pop(0)}")
# plt.plot([x for x in range(self.T)], self.hawk_population_history, label="Hawk Population")
# plt.plot([x for x in range(self.T)], self.total_population_history, label="Total Population")
plt.title(f"Optimal stopping point for 2 agents")
plt.xlabel("Defined stopping point")
plt.ylabel("Percent Explored")
plt.legend()
plt.savefig(f"../search_data_update/2botgraph.png")
plt.show()
```
